```python
import csv

with open('../data/obsolete/AllDataAvocado-transposed.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)
    num_columns = len(header)
    for i, row in enumerate(reader, start=2):  # Start from 2 because we have already read the header
        if len(row) != num_columns:
            print(f"Line {i}: expected {num_columns} columns, but found {len(row)} columns.")
```
